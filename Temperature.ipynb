```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Weather2020.csv')
```

```python
df["Date"] = pd.to_datetime(df['Date'], format="%m/%d/%Y")
```

```python
# Set figure size
plt.figure(figsize=(24,11))

# Get the data
dates = df['Date']
min_temp = (df['Min_TemperatureF']-32) * (5/9)
max_temp = (df['Max_Temperature']-32) * (5/9)
ave_temp = (df['Ave_TemperatureF']-32) * (5/9)

# Create the plot
plt.plot(dates,min_temp,label='Minimum Temperature', linestyle='dashed', color='blue', alpha=0.5)
plt.plot(dates,max_temp,label='Maximum Temperature', linestyle='dashed', color='red', alpha=0.5)
plt.plot(dates,ave_temp,label='Average Temperature', linestyle='dashed', color='green')


# Add a legend
plt.legend(fontsize=14)

# Add a title
plt.title('Temperature over Time, 2021', fontsize=17)

# Add x and y labels
plt.xlabel('Date', fontsize=17)
plt.ylabel('Temperature (Degrees C)', fontsize=17)


plt.xticks(pd.date_range(min(dates), max(dates), freq='7D'), rotation=45, fontsize=14)

# Set y tick marks
plt.yticks(fontsize=14)

# Show the plot
#plt.show()

plt.savefig('Temp2020.png', dpi=300)
```
